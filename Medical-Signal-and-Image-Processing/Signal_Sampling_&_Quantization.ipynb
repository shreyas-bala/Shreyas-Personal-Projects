## **Conceptual Overview Section**

Signal Sampling:
Signal Sampling is like taking snapshots of a continous signal at regular intervals

Quantization:
Quantization is the process of mapping these continuous values to a finite set of discrete values

## **Code Examples Section**



```python
import numpy as np
import matplotlib.pyplot as plt

def generate_signal(frequency, duration, sampling_rate):
    t = np.arange(0, duration, 1/sampling_rate)
    signal = np.sin(2 * np.pi * frequency * t)
    return t, signal

def sample_signal(signal, sampling_rate):
    sampled_signal = signal[::sampling_rate]
    return sampled_signal

def quantize_signal(signal, bits):
    quantization_levels = 2**bits
    quantized_signal = np.round(signal * (quantization_levels - 1)) / (quantization_levels - 1)
    return quantized_signal

# Parameters
frequency = 5  # Frequency of the sine wave
duration = 1  # Duration of the signal in seconds
sampling_rate = 100  # Sampling rate in Hz
quantization_bits = 4  # Number of bits for quantization

# Generate a sine wave signal
time, original_signal = generate_signal(frequency, duration, sampling_rate)

# Sample the signal
sampled_signal = sample_signal(original_signal, 10)

# Quantize the signal
quantized_signal = quantize_signal(sampled_signal, quantization_bits)

plt.figure(figsize=(12, 6))

plt.subplot(3, 1, 1)
plt.plot(time, original_signal, label='Original Signal')
plt.title('Original Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()

plt.subplot(3, 1, 2)
plt.stem(time[::10], sampled_signal, label='Sampled Signal')
plt.title('Sampled Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()

plt.subplot(3, 1, 3)
plt.stem(time[::10], quantized_signal, label='Quantized Signal')
plt.title('Quantized Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()

plt.tight_layout()
plt.show()
```
