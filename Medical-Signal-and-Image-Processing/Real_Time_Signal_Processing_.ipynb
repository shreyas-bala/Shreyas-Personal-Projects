## **Code Examples Section**



```python
import math
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
import time

# Number of samples and sample rate
num_samples = 1024  # Ensure this is a power of 2
sample_rate = 500  # in Hz

# Time vector
time_points = [i/sample_rate for i in range(num_samples)]

#synthetic ECG signal
def synthetic_ecg(t):
    ecg = [math.sin(2.0 * math.pi * 1.0 * i) + 0.5 * math.sin(2.0 * math.pi * 7.0 * i) for i in time_points]
    noise = [np.random.uniform(-0.5, 0.5) for i in time_points]
    ecg = [sum(x) for x in zip(ecg, noise)]
    return ecg

original_signal = synthetic_ecg(time_points)

# Function for real-time signal processing using a moving average filter
def real_time_processing(signal, window_size):
    filtered_signal = np.zeros_like(signal)

    for i in range(len(signal)):
        if i < window_size:
            filtered_signal[i] = np.mean(signal[:i+1])
        else:
            filtered_signal[i] = np.mean(signal[i-window_size+1:i+1])

    return filtered_signal

window_size = 50

# Real-time signal processing with a moving average filter
filtered_signal = real_time_processing(original_signal, window_size)

# Plot original signal
plt.figure(figsize=(10, 4))
plt.subplot(2, 1, 1)
plt.plot(time_points, original_signal, label='Original Signal')
plt.title('Original Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()



# Plot filtered signal
plt.subplot(2, 1, 2)
plt.plot(time_points, filtered_signal, label='Filtered Signal')
plt.title('Real-Time Filtered Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()

# Show plots
plt.tight_layout()
plt.show()
```

## **Solution Development Section**

1. **Solution Explanation:** The implemented algorithm involves real-time signal processing using a moving average filter on a synthetic Electrocardiogram (ECG) signal. The moving average filter, a common technique in signal processing, calculates the average amplitude within a specified window at each time point, effectively smoothing the ECG signal and revealing underlying trends by reducing high-frequency noise.
