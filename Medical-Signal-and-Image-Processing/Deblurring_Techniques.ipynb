## **Conceptual Overview Section**

Inverse Filtering: Inverse filtering is a deblurring technique that attempts to recover the original image by estimating the inverse of the degradation process, assuming knowledge of the blur kernel.

Wiener Filtering: Wiener deconvolution is a method for image deblurring that minimizes the impact of noise by employing a statistical approach, utilizing knowledge of the signal-to-noise ratio (SNR).

## **Code Examples Section**



INVERSE FILTERING

```python

import cv2
import numpy as np
from scipy.signal import convolve2d
from google.colab.patches import cv2_imshow

# Load the image
image_path = "input_image.jpg"
original_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Define a Gaussian blur kernel (you can adjust the kernel size and standard deviation)
blur_kernel_size = 3
blur_kernel = cv2.getGaussianKernel(blur_kernel_size, -1)
blur_kernel = np.outer(blur_kernel, blur_kernel.transpose())

# Apply Gaussian blur to the image
blurred_image = cv2.filter2D(original_image, -1, blur_kernel)

# Display the original and blurred images
cv2_imshow(original_image)
cv2_imshow(blurred_image)
cv2.waitKey(0)
cv2.destroyAllWindows()

# Deblur using inverse filtering
inverse_blur_kernel = np.linalg.pinv(blur_kernel)
deblurred_image = cv2.filter2D(blurred_image, -1, inverse_blur_kernel)

# Display the deblurred image
cv2_imshow(deblurred_image)
cv2.waitKey(0)
cv2.destroyAllWindows()

```

WIENER FILTERING

```python
import cv2
import numpy as np
from scipy.signal import convolve2d

# Load the image in grayscale
image_path = "input_image.jpg"
original_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Define a Gaussian blur kernel
blur_kernel_size = 3
blur_kernel = cv2.getGaussianKernel(blur_kernel_size, 0)
blur_kernel = np.outer(blur_kernel, blur_kernel.transpose())

# Apply Gaussian blur to the grayscale image
blurred_image = cv2.filter2D(original_image, -1, blur_kernel)

# Display the original and blurred images
cv2_imshow(original_image)
cv2_imshow( blurred_image)
cv2.waitKey(0)
cv2.destroyAllWindows()

# Wiener deconvolution
# Adjust SNR
snr = 1  # Signal-to-noise ratio
wiener_filter = np.conj(blur_kernel) / (np.abs(blur_kernel) ** 2 + 1 / snr)

# Apply Wiener deconvolution
deblurred_image_wiener = cv2.filter2D(blurred_image, -1, wiener_filter)

# Display the deblurred image using Wiener deconvolution
cv2_imshow(deblurred_image_wiener)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
